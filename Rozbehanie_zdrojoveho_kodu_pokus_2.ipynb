```
import numpy 
import math
import cv2
original = cv2.imread('./results/test_original/final/10228_00_300_gt.png')
contrast = cv2.imread('./results/test_original/final/10228_00_300_gt.png',1)
def psnr(img1, img2):
    mse = numpy.mean((img1 - img2) ** 2)
    if mse == 0:
      return 100
    PIXEL_MAX = 255.0
    return 20 * math.log10(PIXEL_MAX / math.sqrt(mse))

d = psnr(original,contrast)
print(d)
```
